```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
rng.exponential(scale=1.0, size=(4, 4))
```

```python
rng = np.random.default_rng()
rng.exponential(scale=1.0, size=(1000))
```

```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue",
label="rand")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red",
label="randn")
plt.hist(rng.exponential(scale=1.0, size=(1000)), density=True, bins=100, histtype="step", color="green",
label="expo")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True,
           hue='Stage')
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(data=df)
```

```python
# Preprocess DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# New boxplot using stats_df
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=stats_df)
```
